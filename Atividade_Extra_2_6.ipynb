## **Exemplo 5.2**
**Enunciado:** Cálculo da quantidade de bits necessários para uma memória cache de 32KB, linhas de 8 bytes e MP de 16MB (Mapeamento Direto).

```python
def calcular_exemplo_5_2():
    # Dados
    cache_cap_kb = 32
    linha_b = 8
    mp_cap_mb = 16

    # 1. Total de bits de dados (Td)
    td_bits = cache_cap_kb * 1024 * 8

    # 2. Quantidade de linhas
    num_linhas = (cache_cap_kb * 1024) // linha_b

    # 3. Largura do campo Tag
    # MP / Largura da linha = Total de blocos
    total_blocos = (mp_cap_mb * 1024 * 1024) // linha_b
    tag_bits = (total_blocos // num_linhas).bit_length() - 1

    # 4. Total de bits de Tag (Tt)
    tt_bits = num_linhas * tag_bits

    # Total
    total_bits = td_bits + tt_bits

    print(f"Td (Dados): {td_bits} bits")
    print(f"Linhas: {num_linhas}")
    print(f"Tag: {tag_bits} bits")
    print(f"Tt (Tags): {tt_bits} bits")
    print(f"Total: {total_bits} bits (~{total_bits/1024:.0f} Kbits)")

calcular_exemplo_5_2()
```

## **Exemplo 5.3**
**Enunciado:** Formato do endereço para MP de 64MB, cache com 2K linhas de 16 bytes (Mapeamento Direto).

```python
import math

def calcular_exemplo_5_3():
    mp_mb = 64
    linhas = 2048 # 2K
    larg_linha = 16

    # Bits totais do endereço
    total_addr_bits = int(math.log2(mp_mb * 1024 * 1024))

    # Campo Byte
    byte_bits = int(math.log2(larg_linha))

    # Campo Linha
    linha_bits = int(math.log2(linhas))

    # Campo Tag
    tag_bits = total_addr_bits - linha_bits - byte_bits

    print(f"Total Endereço: {total_addr_bits} bits")
    print(f"Formato: [Tag: {tag_bits} | Linha: {linha_bits} | Byte: {byte_bits}]")

calcular_exemplo_5_3()
```

## **Exemplo 5.4**
**Enunciado:** Localizar a linha do endereço hexadecimal `3FC92B6` (Cache 128KB, Bloco 32B).

```python
def calcular_exemplo_5_4():
    addr_hex = "3FC92B6"
    cache_size = 128 * 1024
    block_size = 32

    # Conversão para binário (28 bits)
    addr_bin = bin(int(addr_hex, 16))[2:].zfill(28)

    # Cálculo dos campos
    byte_bits = int(math.log2(block_size)) # 5
    num_linhas = cache_size // block_size # 4096
    linha_bits = int(math.log2(num_linhas)) # 12

    # Divisão: [Tag (11) | Linha (12) | Byte (5)]
    tag = addr_bin[:11]
    linha = addr_bin[11:11+12]
    byte = addr_bin[23:]

    print(f"Endereço Binário: {addr_bin}")
    print(f"Tag: {tag} | Linha: {linha} | Byte: {byte}")
    print(f"Linha buscada (bin): {linha}")

calcular_exemplo_5_4()
```

## **Exemplo 5.5**
**Enunciado:** Cálculo da quantidade de bits necessários para uma memória cache de 32KB, linhas de 8 bytes e MP de 16MB (Mapeamento Associativo).

Neste exemplo, o foco é a soma dos bits de dados com os bits do campo **bloco** (tag) presente em cada linha.

```python
def calcular_exemplo_5_5():
    # Dados do problema
    cache_cap_kb = 32
    linha_b = 8
    mp_cap_mb = 16

    # 1. Total de bits para a parte de dados (Td)
    td_bits = cache_cap_kb * 1024 * 8

    # 2. Quantidade de linhas na cache
    num_linhas = (cache_cap_kb * 1024) // linha_b

    # 3. Largura do campo Bloco (Tag no associativo completo)
    # Quantidade de blocos na MP = Capacidade MP / Largura do bloco
    total_blocos = (mp_cap_mb * 1024 * 1024) // linha_b
    # A largura do campo bloco (bits necessários para endereçar cada bloco)
    largura_bloco_bits = int(math.log2(total_blocos))

    # 4. Total de bits do campo bloco (Tb)
    tb_bits = num_linhas * largura_bloco_bits

    # 5. Total Geral (T)
    total_geral_bits = td_bits + tb_bits

    print(f"--- Resolução Exemplo 5.5 ---")
    print(f"Td (Dados): {td_bits:,} bits")
    print(f"Quantidade de Linhas: {num_linhas:,} (2^12)")
    print(f"Total de Blocos na MP: {total_blocos:,} (2^21)")
    print(f"Largura do Campo Bloco: {largura_bloco_bits} bits")
    print(f"Tb (Total de bits de bloco): {tb_bits:,} bits")
    print(f"Total de bits na Cache (T): {total_geral_bits:,} bits")
    print(f"Equivalente a: {total_geral_bits / 1024:.2f} Kbits")

calcular_exemplo_5_5()
```

## **Exemplo 5.6 (Resolução Separada)**
**Enunciado:** Considere uma MP com 64MB de capacidade associada a uma memória cache que possui 2K linhas, cada uma com largura de 16 bytes. Determine o formato do endereço para mapeamento associativo completo.

```python
import math

def resolucao_exemplo_5_6_separada():
    # Dados
    mp_cap_mb = 64
    cache_linhas = 2048  # 2K linhas
    largura_linha_b = 16

    # 1. Bits totais do endereço (MP)
    total_addr_bits = int(math.log2(mp_cap_mb * 1024 * 1024))

    # 2. Campo Byte (identifica o byte dentro do bloco/linha)
    # log2(16 bytes) = 4 bits
    byte_bits = int(math.log2(largura_linha_b))

    # 3. Campo Bloco (identifica qual bloco da MP está na cache)
    # No mapeamento associativo completo: Bits do Bloco = Total - Byte
    bloco_bits = total_addr_bits - byte_bits

    # 4. Cálculo do total de blocos para conferência
    total_blocos = (mp_cap_mb * 1024 * 1024) // largura_linha_b

    print(f"--- Resolução Exemplo 5.6 ---")
    print(f"Capacidade da MP: {mp_cap_mb} MB ({total_addr_bits} bits de endereço)")
    print(f"Largura da Linha: {largura_linha_b} bytes")
    print(f"Total de blocos na MP: {total_blocos:,} (2^{bloco_bits})")
    print(f"\nFormato do Endereço (Associativo Completo):")
    print(f"- Bloco: {bloco_bits} bits")
    print(f"- Byte: {byte_bits} bits")
    print(f"\nEstrutura: [ Bloco ({bloco_bits} bits) | Byte ({byte_bits} bits) ]")

resolucao_exemplo_5_6_separada()
```

## **Exemplo 5.7**
**Enunciado:** Determinar o valor binário do campo **bloco** para o endereço `3FC92B6` em uma cache com blocos de 32 bytes e mapeamento associativo completo.

```python
def calcular_exemplo_5_7():
    addr_hex = "3FC92B6"
    block_size = 32

    # 28 bits (7 hex * 4)
    addr_bin = bin(int(addr_hex, 16))[2:].zfill(28)

    # Byte bits = log2(32) = 5
    byte_bits = 5
    bloco_bits = 28 - byte_bits

    bloco_val = addr_bin[:bloco_bits]
    byte_val = addr_bin[bloco_bits:]

    print(f"Endereço Hex: {addr_hex}")
    print(f"Endereço Binário: {addr_bin}")
    print(f"Campo Bloco ({bloco_bits} bits): {bloco_val}")
    print(f"Campo Byte ({byte_bits} bits): {byte_val}")

calcular_exemplo_5_7()
```

## **Exemplo 5.8 (Resolução Separada)**
**Enunciado:** Cálculo da quantidade de bits necessários para uma determinada memória cache, que funciona com mapeamento por conjunto de quatro. Considere um sistema de computação com uma memória cache de 32KB de capacidade, constituída de linhas com 8 bytes de largura e conjunto de 4. A MP possui uma capacidade de 16MB.

```python
import math

def resolucao_exemplo_5_8_separada():
    # 1. Dados iniciais
    cache_cap_kb = 32
    linha_b = 8
    set_size = 4
    mp_cap_mb = 16

    # 2. Total de bits de dados (Td)
    # 32KB * 1024 * 8 bits/byte
    td_bits = cache_cap_kb * 1024 * 8

    # 3. Cálculo da quantidade de linhas
    num_linhas = (cache_cap_kb * 1024) // linha_b

    # 4. Cálculo da quantidade de conjuntos (C)
    num_sets = num_linhas // set_size

    # 5. Cálculo do campo Tag
    # Total de blocos na MP = 16MB / 8B
    total_blocos_mp = (mp_cap_mb * 1024 * 1024) // linha_b
    # Blocos por conjunto = Total de blocos / Quantidade de conjuntos
    blocos_por_conjunto = total_blocos_mp // num_sets
    # Largura do Tag = log2(blocos por conjunto)
    tag_bits = int(math.log2(blocos_por_conjunto))

    # 6. Total de bits de Tag (Tt)
    # Cada linha da cache possui uma tag
    tt_bits = num_linhas * tag_bits

    # 7. Total Geral (T)
    total_bits = td_bits + tt_bits

    print(f"--- Resolução Exemplo 5.8 ---")
    print(f"Td (Dados): {td_bits:,} bits")
    print(f"Linhas: {num_linhas:,} | Conjuntos: {num_sets:,}")
    print(f"Largura da Tag: {tag_bits} bits")
    print(f"Tt (Tags): {tt_bits:,} bits")
    print(f"Total Geral (T): {total_bits:,} bits")
    print(f"Equivalente a: {total_bits / 1024:.2f} Kbits")

resolucao_exemplo_5_8_separada()
```

## **Exemplo 5.9 (Resolução Separada)**
**Enunciado:** Considere um sistema de computação com uma MP de 64MB e uma memória cache de 32KB. A cache trabalha com mapeamento associativo por conjunto de 4 e suas linhas têm largura de 16 bytes. Determine o formato do endereço.

```python
import math

def resolucao_exemplo_5_9_separada():
    # 1. Dados iniciais
    mp_cap_mb = 64
    cache_cap_kb = 32
    linha_b = 16
    set_size = 4

    # 2. Bits totais do endereço (log2 da capacidade da MP)
    total_addr_bits = int(math.log2(mp_cap_mb * 1024 * 1024))

    # 3. Campo Byte (identifica o byte dentro da linha/bloco)
    # log2(16) = 4 bits
    byte_bits = int(math.log2(linha_b))

    # 4. Cálculo do campo Conjunto
    # Primeiro, calculamos o total de linhas: 32KB / 16B = 2048 linhas
    total_linhas = (cache_cap_kb * 1024) // linha_b
    # Quantidade de conjuntos: 2048 / 4 = 512 conjuntos
    num_sets = total_linhas // set_size
    # Bits de Conjunto: log2(512) = 9 bits
    set_bits = int(math.log2(num_sets))

    # 5. Cálculo do campo Tag
    # Tag = Total - Conjunto - Byte
    tag_bits = total_addr_bits - set_bits - byte_bits

    print(f"--- Resolução Exemplo 5.9 ---")
    print(f"Capacidade MP: {mp_cap_mb} MB ({total_addr_bits} bits de endereço)")
    print(f"Capacidade Cache: {cache_cap_kb} KB")
    print(f"Linhas: {total_linhas} | Conjuntos: {num_sets}")
    print(f"\nFormato do Endereço (Associativo por Conjunto):")
    print(f"- Tag: {tag_bits} bits")
    print(f"- Conjunto: {set_bits} bits")
    print(f"- Byte: {byte_bits} bits")
    print(f"\nEstrutura: [ Tag ({tag_bits}) | Conjunto ({set_bits}) | Byte ({byte_bits}) ]")

resolucao_exemplo_5_9_separada()
```

## **Exemplo 5.10 (Resolução Separada)**
**Enunciado:** Seja uma MP constituída de blocos com largura de 32 bytes, associada a uma cache com 64KB; a cache usa mapeamento por conjunto de 4. Em dado instante o processador realiza um acesso, colocando o seguinte endereço (expresso em algarismos hexadecimais): 3FC92B6. Determine qual deverá ser o valor binário do conjunto que será localizado pelo sistema de controle da cache.

```python
import math

def resolucao_exemplo_5_10_detalhada():
    # 1. Configurações
    addr_hex = "3FC92B6"
    cache_size = 64 * 1024  # 64 KB
    block_size = 32         # 32 Bytes
    set_size = 4            # Conjunto de 4
    mp_total_bits = 28      # 7 dígitos hex * 4 bits

    # 2. Conversão do endereço para binário
    addr_bin = bin(int(addr_hex, 16))[2:].zfill(mp_total_bits)

    # 3. Cálculos dos campos
    # Bits de Byte: log2(32) = 5 bits
    byte_bits = int(math.log2(block_size))

    # Quantidade de linhas: 64KB / 32B = 2048 linhas
    total_linhas = cache_size // block_size

    # Quantidade de conjuntos: 2048 / 4 = 512 conjuntos
    num_sets = total_linhas // set_size

    # Bits de Conjunto: log2(512) = 9 bits
    set_bits = int(math.log2(num_sets))

    # Bits de Tag: 28 - 9 - 5 = 14 bits
    tag_bits = mp_total_bits - set_bits - byte_bits

    # 4. Extração dos valores
    # Formato: [ Tag (14) | Conjunto (9) | Byte (5) ]
    tag_val = addr_bin[:tag_bits]
    set_val = addr_bin[tag_bits : tag_bits + set_bits]
    byte_val = addr_bin[tag_bits + set_bits :]

    print(f"Endereço Hexadecimal: {addr_hex}")
    print(f"Endereço Binário Completo: {addr_bin}")
    print(f"\nDistribuição dos Campos:")
    print(f"- Tag ({tag_bits} bits): {tag_val}")
    print(f"- Conjunto ({set_bits} bits): {set_val}")
    print(f"- Byte ({byte_bits} bits): {byte_val}")
    print(f"\nResultado: O valor binário do conjunto é {set_val}")

resolucao_exemplo_5_10_detalhada()
```
